```python
#b
import pymc as pm
import numpy as np
import matplotlib.pyplot as plt
import arviz as az

np.random.seed(16)

x = np.random.randint(1,7)
crosii = 0
calbastre = 0
cnegre = 0

for i in range (1, 30001):
    rosii = 3
    albastre = 4
    negre = 2
    x = np.random.randint(1,7)
    if x == 2:
        negre = negre + 1
    elif x == 3:
        negre = negre + 1
    elif x == 5:
        negre = negre + 1
    elif x == 6:
        rosii = rosii +1
    else: 
        albastre = albastre + 1
    y = np.random.rand(1) * 10
    rosii = rosii
    albastre = rosii + albastre
    negre = negre + albastre
    if(y <= rosii):
        crosii = crosii +1
    elif(y <= albastre):
        calbastre = calbastre + 1
    else:
        cnegre = cnegre +1

print(crosii, calbastre, cnegre)

probRosu = crosii / (crosii + calbastre + cnegre)
print("probabilitatea de Rosu: ", probRosu)

culori = (crosii, calbastre, cnegre)
labels = 'rosu', 'albastru', 'negru'
plt.style.use('_mpl-gallery-nogrid')
fig, ax = plt.subplots()
ax.pie(culori, radius=3, center=(4, 4), labels=labels, autopct='%1.1f%%')

plt.show()
```
